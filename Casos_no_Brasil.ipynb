```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('tabela de casos.csv',low_memory=False,delimiter=';', encoding='UTF-8', decimal=',')

df
colunas =[]

x = df.columns
x = x.drop("anos/doenca")
linha = df.head(10)
total = []
for coluna in linha:
    if coluna != "anos/doenca":
        total.append(linha[coluna][9])
fig, ax = plt.subplots(figsize=(20, 10), dpi=200)
plt.bar(x, total)

plt.title('5 doenças zoonoses com mais casos registrados no Brasil (2010-2018)', fontsize=17)
plt.xlabel('Doenças zoonoses', fontsize=17)
plt.ylabel('Casos', fontsize=17)
plt.legend(("Casos Totais",),fontsize=20)
plt.ylim(100,100000)
plt.show()
```

```python
x = df.iloc[:9, 0]
salmonelose = df.iloc[:9, 2]
leptospirose = df.iloc[:9, 3]
toxoplasmose = df.iloc[:9, 4]
brucelose = df.iloc[:9, 5]

fig, ax = plt.subplots(figsize=(20, 8), dpi=200)

barWidth =0.20

r1 = np.arange(len(salmonelose))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

plt.bar(r2, salmonelose,barWidth)
plt.bar(r1, leptospirose,barWidth)
plt.bar(r3, toxoplasmose,barWidth)
plt.bar(r4, brucelose,barWidth)

plt.xticks([r + barWidth for r in range(len(salmonelose))],x)   
plt.title('Comparando as doenças zoonoses por ano\n *Pardindo da segunda com mais casos', fontsize=17)
plt.xlabel('Ano', fontsize=17)
plt.ylabel('Casos', fontsize=17)
plt.legend(("Casos de salmonelose","Casos de leptospirose","Casos de toxoplasmose","Casos de brucelose"), fontsize=15)
plt.ylim(0,5000)
plt.show()
```
